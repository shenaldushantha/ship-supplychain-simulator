## plot helper

```python
def plot_gantt(tasks, order, cp, title='schedule', save_path=None):
    fig = plt.figure(figsize=(10,5)); ax = plt.gca()
    for idx, tid in enumerate(order):
        t = tasks[tid]; ax.barh(idx, t.duration, left=t.es)
        ax.text(t.es+t.duration+0.5, idx, f"{t.duration:.0f}d", va='center', fontsize=9)
    ax.set_yticks(range(len(order))); ax.set_yticklabels([f"{tid}{'*' if tid in cp else ''}" for tid in order])
    ax.set_xlabel('days'); ax.set_title(title); fig.tight_layout()
    if save_path: fig.savefig(save_path, dpi=200, bbox_inches='tight')
    plt.show(); plt.close(fig)
```

this just draws a simple gantt chart so i can see the tasks and the critical ones.

## base chart

```python
plot_gantt(tasks, order, cp, title='base schedule (critical *)', save_path='../data/schedule_export.png')
```

this shows the main schedule, everything on time. the star * marks critical tasks.
